```python
from policyengine_us import Microsimulation

sim = Microsimulation(dataset="enhanced_cps_2023")
```

```python
agi = sim.calculate("adjusted_gross_income")

import numpy as np
import pandas as pd

bins = [0, 1, 2.5e4, 5e4, 1e5, 2e5, 5e5, 1e6, np.inf]

lower_bounds = bins[:-1]
agi_totals = []
agi_counts = []

for i in range(len(bins) - 1):
    agi_counts.append(
        agi[(agi >= lower_bounds[i]) * (agi < bins[i + 1])].count()
    )
    agi_totals.append(
        agi[(agi >= lower_bounds[i]) * (agi < bins[i + 1])].sum()
    )

df = pd.DataFrame(
    {"lower_bound": lower_bounds, "total": agi_totals, "count": agi_counts}
)
df["lower_bound"] = df["lower_bound"].apply(lambda x: f"${x/1e3:.1f}k")
df["count"] = df["count"].apply(lambda x: f"{x/1e6:.1f}m")
df["total"] = df["total"].apply(lambda x: f"${x/1e12:,.1f}tr")
df
```
